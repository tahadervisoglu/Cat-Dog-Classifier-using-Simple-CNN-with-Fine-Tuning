```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
base_dir='cats_and_dogs_filtered'
train_dir= os.path.join(base_dir,'train')
val_dir = os.path.join(base_dir, 'validation')
```

```python
img_height, img_width = 150, 150
batch_size = 32
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
val_generator= val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # binary classification
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=40,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)
```

```python
model.save('my_model.h5')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

# 1. Kullanıcıdan dosya yüklemesini iste
uploaded = files.upload()

# 2. Yüklenen ilk dosyanın ismini al
filename = list(uploaded.keys())[0]

# 3. Görseli aç, RGB formatına çevir
img = Image.open(filename).convert('RGB')
img_cv = np.array(img)
img_cv = cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR)

# 4. Görseli modele uygun boyuta getir (150x150)
img_resized = cv2.resize(img_cv, (150, 150))
img_array = img_resized / 255.0
img_array = np.expand_dims(img_array, axis=0)

# 5. Model ile tahmin yap
prediction = model.predict(img_array)[0][0]
label = "Dog" if prediction > 0.5 else "Cat"

# 6. Sonucu görsele yaz
cv2.putText(img_cv, label, (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 3)

# 7. Görseli RGB’ye çevir ve göster
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title(f"Prediction: {label}")
plt.axis('off')
plt.show()
```
